```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
```

```python
hearing = pd.read_csv('./data/hearing_test.csv')
airlines = pd.read_csv('./data/Invistico_Airline.csv')
mushroom = pd.read_csv('./data/secondary_data.csv')
weather = pd.read_csv('./data/weather_forecast_data.csv')
```

```python
from sklearn.model_selection import train_test_split

X = hearing.drop('test_result', axis=1)
y = hearing['test_result']


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)


x_train.shape
```

```python
log_r = LogisticRegression(max_iter=500)

log_r.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_preds_log_r = log_r.predict(x_test)

cm = confusion_matrix(y_test, y_preds_log_r, labels=[0, 1])
ConfusionMatrixDisplay(cm ).plot()
```
